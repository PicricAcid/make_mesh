```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

aa = 3 # length in x-direction
bb = 1 # length in y-direction
nn = 4 # division number in x-direction
nn = 2*nn
mm = 2 # division number in y-direction
mm = 2*mm
x0 = 0 # x-coordinate at left bottom of the domain
y0 = 0 # y-coordinate at left bottom of the domain

a = 0.1

npoin = (nn + 1)*(mm + 1)
nele = int((nn*mm)/4)
node = np.zeros([8, nele], dtype=np.int)
x = np.zeros([2, npoin], dtype=np.float64)

def func(x, maxy):
    if(x < a):
        y = np.sqrt(a**2 - x**2)
    else:
        y = 0
    
    return y

k = -1
for i in range(0, mm+1):
    if(i%2 == 1):
        for j in range(0, nn+1):
            if(j%2 == 0):
                k = k+1
                x[0, k] = aa/float(nn)*float(j) + x0
                f = func(x[0, k], bb)
                x[1, k] =  (bb - f)/float(mm)*float(i) + f
    else:
        for j in range(0, nn+1):
            k = k+1
            x[0, k] = aa/float(nn)*float(j) + x0
            f = func(x[0, k], bb)
            x[1, k] = (bb - f)/float(mm)*float(i) + f

x = np.array([x[0, :k+1], x[1, :k+1]])
npoin = k+1
    
ne = -1
for k in range(0, int(mm/2)):
    kk = k + 1
    i0 = 1 + (nn + nn/2 + 2)*kk
    
    for j in range(0, int(nn/2)):
        i = i0 + 2*j
        ne = ne + 1
        #  0 1 2
        #  7    3
        # 6 5 4
        node[0, ne] = i
        node[1, ne] = i + 1
        node[2, ne] = i + 2
        node[3, ne] = node[2, ne] - (2 + j + nn/2) 
        node[4, ne] = node[3, ne] - (nn - j)
        node[5, ne] = node[4, ne] - 1
        node[6, ne] = node[5, ne] - 1
        node[7, ne] = node[5, ne] + (nn - j)


face = np.zeros((0, 9))
for ne in range(0, nele):
    ss = []
    for i in range(0, 8):
        ss.append(node[i, ne])
    ss.append(ne+1)
    ss = np.array(ss)
    face = np.r_[face, ss.reshape(1, -1)]

node = np.zeros((0, 3))
for nd in range(0, npoin):
    ss = []
    for i in range(0,2):
        ss.append(x[i, nd])
    ss.append(nd+1)
    ss = np.array(ss)
    node = np.r_[node, ss.reshape(1, -1)]

xx = node[:, 0]
yy = node[:, 1]
plt.axes().set_aspect('equal')
plt.scatter(xx, yy)

print(face)

for i in range(np.shape(node)[0]):
    plt.text(node[i, 0], node[i, 1], node[i, 2])
    
path = "./mesh.txt"
with open(path, mode='w') as f:
    f.write("{}\n".format(npoin))
    for i in range(np.shape(node)[0]):
        f.write("{0[0]: <.4f} {0[1]: <.4f}\n".format(node[i, :]))
    
    f.write("{}\n".format(nele))
    for i in range(np.shape(face)[0]):
        f.write("{0[0]: <.0f} {0[1]: <.0f} {0[2]: <.0f} {0[3]: <.0f} {0[4]: <.0f} {0[5]: <.0f} {0[6]: <.0f} {0[7]: <.0f}\n".format(face[i, :]))
```
